```python
import pandas as pd
import numpy as np
from qgrid import show_grid as exl
from ast import literal_eval

from scipy.sparse import hstack,csr_matrix
```

```python
df = pd.read_csv('movies_metadata.csv')[:10000]
df.id = df.id.astype('int64')
df.head()
```

```python
keyword = pd.read_csv('keywords.csv')[:10000]
keyword.head()
```

```python
credit = pd.read_csv('credits.csv')[:10000]
credit.head()
```

```python
keyword.info()
```

```python
keyword['kw'] = keyword.keywords.apply(lambda row:[i['name'].lower().replace(' ', '_') for i in literal_eval(row)[:3]])
```

```python
keyword
```

```python
for i in literal_eval(credit.cast[0]):
    print(i['name'].lower().replace(' ', '_'))
```

```python
credit['cast'] = credit.cast.apply(lambda row: [i['name'].lower().replace(' ', '_') for i in literal_eval(row)])
```

```python
df = df.merge(credit, on='id')
```

```python
df = df.merge(keyword, on='id')
```

```python
df = df[['genres','popularity','overview','production_companies','title','cast','kw', 'id']]
df.head()
```

```python
df.shape
```

```python
df.genres = df.genres.apply(lambda row: [i['name'].lower() for i in literal_eval(row)])
```

```python
df.production_companies = df.production_companies.apply(lambda row: [i['name'].lower().replace(' ', '_') for i in literal_eval(row)])
```

```python
df.overview.fillna('',inplace=True)
df.overview = df.overview.apply(lambda row: row.lower())
```

```python
df.popularity = df.popularity.astype('int64')
```

```python
df.head()
```

```python
def create_final_text_dataset(row):
    return ' '.join(row.genres)+' '.join(row.production_companies)+' '+' '.join(row.cast)+' '+' '.join(row.kw)#+' '+row.overview
```

```python
df['soup'] = df.apply(create_final_text_dataset, axis=1)
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_features=500)
```

```python
tfidf_mat = tfidf.fit_transform(df.soup)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# similarity = cosine_similarity(final_text_tfidf_mat)  #This is not working so do not use this thing
similarity = cosine_similarity(tfidf_mat)
```

```python
def recommend(title='Toy Story'):
    idx = df.loc[df.title==title].index
    print(idx)
    similarity_score = similarity[idx][0]
    similar_movie_idx = sorted(enumerate(similarity_score), key=lambda x:x[1], reverse=True)[:11]#[1:11]
    similar_movies = [df.iloc[idx].title for idx, score in similar_movie_idx]
    similar_movies_genre = [df.iloc[idx].genres for idx, score in similar_movie_idx]
    similar_movies_popularity = [df.iloc[idx].popularity for idx, score in similar_movie_idx]    
    similar_movies_cast = [df.iloc[idx].cast for idx, score in similar_movie_idx]        
    similar_movies_kw = [df.iloc[idx].kw for idx, score in similar_movie_idx]
    similar_movies_pc = [df.iloc[idx].production_companies for idx, score in similar_movie_idx]       
    similar_movies_score = [score for idx, score in similar_movie_idx]
    
    return pd.DataFrame({
        "Movie":similar_movies,
        "Genre":similar_movies_genre,
        "Score":similar_movies_score,
        "Popularity":similar_movies_popularity,
        'Cast':similar_movies_cast,
        'Keyword':similar_movies_kw,
        "Production Company":similar_movies_pc
    })
```

```python
recommend("The Lion King")
```
